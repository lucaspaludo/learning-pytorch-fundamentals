```python
import torch
import numpy as np
```

```python
tensorA = torch.tensor([1, 2, 3, 4])
tensorB = torch.tensor([5, 6, 7, 8])
```

```python
tensorA + tensorB
```

```python
torch.add(tensorA, tensorB)
```

```python
tensorB - tensorA
```

```python
torch.sub(tensorB, tensorA)
```

```python
tensorA * tensorB
```

```python
torch.mul(tensorA, tensorB)
```

```python
tensorB / tensorA
```

```python
torch.div(tensorB, tensorA)
```

```python
tensorB % tensorA
```

```python
torch.remainder(tensorB, tensorA)
```

```python
tensorA ** tensorB
```

```python
torch.pow(tensorA, tensorB)
```

```python
tensorA.add(tensorB)
```
